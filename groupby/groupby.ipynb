```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
```

```python
drinks.head()
```

```python
drinks.beer_servings.mean()
```

```python
drinks['beer_servings'].mean()
```

groupby function is to split the data into groups on some or similar criteria

```python
##average beer servings by continents
drinks.groupby('continent').beer_servings.mean()
##groupby basically filtering the frame by continent, then taking the servings' mean!
```

```python
#similar action to groupby
#this one choosing only the africa from the column
# drinks[drinks.continent == 'Africa']
#this statement is the same as groupby method we used above
drinks[drinks.continent == 'Africa'].beer_servings.mean()
#possible to use any other aggregated functions
```

```python
#aggregate function allows us to use multiple statistical functions to use at once
drinks.groupby('continent').beer_servings.agg(['count', 'max', 'min', 'mean'])
```

```python
c = drinks.groupby(['continent'], sort = True)
c
```

```python
#to print the first entries in all groups
c.first()
```

```python
#Iterate an element of group
#print out the all elements that grouped
for continent, drinks in c:
    print(continent)
```

```python
#in order to select a single group we use get_group()
#grupladiktan sonra o gruptan sadece bir taneni secmek icin
#
c.get_group("Europe")
#this is same as "SELECT * FROM "
```
