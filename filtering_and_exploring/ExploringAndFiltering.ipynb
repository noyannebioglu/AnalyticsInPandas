```python
#Will be using some major pandas functions code to explore and filter data
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("employees.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
#renaming the column names to reach them easily
mapping = {"First Name" : "fn", "Gender" : "gender", "Start Date" : "start_date", "Last Login Time" : "last_login", "Salary" : "salary", "Bonus %" : "bonus", "Team" : "team"}
df = df.rename(mapping, axis=1)
```

```python
#exploring the null values
#selecting all rows with NAN under a single DF column
df[df["fn"].isna()]
#there are 67 NULL values in "fn" column
```

```python
#Selecting all rows with NAN under the entire DF
#the total of 236 NAN values exist DF
df[df.isna().any(axis=1)]
```

```python
#Exploring the salaries in company more than 100K
salary = df[df["salary"] > 100000]
salary.describe()
```

```python
# print('rows')
# print(df[df == 0].count(axis=1)/len(df.columns))
# print('cols')
# print(df[df == 0].count(axis=0)/len(df.index))
df[df == 0].count(axis=1)/len(df.columns)
```

```python
#Replacing the NAN values in columns with string values
df["gender"].fillna("no gender", inplace=True)
df["fn"].fillna("unknown name", inplace=True)
df["team"].fillna("unknown team", inplace=True)
```

```python
df.info()
```

```python
#Experience the effect of given values to NAN rows
#to see the variance after replacing them with 0 number. 
data = {'first_set': [1,2,3,4,5,np.nan,6,7,np.nan,np.nan,8,9,10,np.nan],
        'second_set': ['a','b',np.nan,np.nan,'c','d','e',np.nan,np.nan,'f','g',np.nan,'h','i']
        }
df_numpy = pd.DataFrame(data,columns=['first_set','second_set'])
print (df_numpy)
```

```python
#mean result before replacing
df_numpy["first_set"].mean()
```

```python
df_numpy["first_set"].fillna(0, inplace=True)
```

```python
#mean result after replacing the NAN values with zero
df_numpy["first_set"].mean()
```

```python
#it is good practice to drop or leave NAN values to avoid variations
```

```python
df_numpy["first_set"].mean()
```

###### ---------------------------------------------------------------------------------
## Query()
#### Helps us to filter a DF based on a condition or to apply a mask to get certain values. 

```python
#doing the same thing with query() function
with_query_function = df.query('team == "Marketing" & salary > 100000')
with_query_function.head()
```

```python
#Selecting values between 40K and 100K, and displaying 8th of them
df.query('40000 < salary < 100000')[:8]
```

```python
tilda_employees_over100k_in_marketing = df[~(df.team == "Marketing") & (df.salary > 100000)]
```

```python
tilda_employees_over100k_in_marketing.info()
```

## Loc and iloc
#### Two functions are used to select rows and columns
- iloc : returns pandas series when one row is selected: 
- loc : select by labels. Labels are the column names. 

```python
#Selecting 10th to 15th rows from "gender" and "salary" labels
df.loc[10:15,["gender", "salary"]]
```

```python
#selecting first 10 rows and 4 columns
df.iloc[:10, 0:4]
```
